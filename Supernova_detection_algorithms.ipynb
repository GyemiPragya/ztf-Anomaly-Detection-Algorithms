```python
import ast
import numpy as np

def to_list(val):
    # If already a list or numpy array, convert to list
    if isinstance(val, list):
        return val
    if isinstance(val, np.ndarray):
        return val.tolist()
    # If it's a string, try to evaluate it
    if isinstance(val, str):
        try:
            return ast.literal_eval(val)
        except Exception:
            return []  # or np.nan, depending on your preference
    # If it's something else, return as is or empty list
    return []
```

```python
def extract_features(group):
    return pd.Series({
        'mean_mag': group['mag'].mean(),
        'std_mag': group['mag'].std(),
        'min_mag': group['mag'].min(),
        'max_mag': group['mag'].max(),
        'mag_range': group['mag'].max() - group['mag'].min(),
        'skew_mag': group['mag'].skew(),
        'kurtosis_mag': group['mag'].kurtosis(),
        'num_points': group['mag'].count()
    })

# Exclude the grouping column from the group passed to the function
features_df = (
    long_df.groupby('objectid', group_keys=False)
    .apply(lambda g: extract_features(g))
    .reset_index()
)
print(features_df.head())
```

```python
import torch
import torch.nn as nn

class Autoencoder(nn.Module):
    def __init__(self, input_dim):
        super(Autoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 16),
            nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.Linear(16, input_dim)
        )
    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

# Prepare data
X_tensor = torch.tensor(X_scaled, dtype=torch.float32)

# Model, loss, optimizer
model = Autoencoder(input_dim=X_scaled.shape[1])
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Training loop
for epoch in range(50):
    optimizer.zero_grad()
    outputs = model(X_tensor)
    loss = criterion(outputs, X_tensor)
    loss.backward()
    optimizer.step()

# Anomaly scores
with torch.no_grad():
    recon = model(X_tensor)
    mse = ((X_tensor - recon) ** 2).mean(dim=1).numpy()
threshold = np.percentile(mse, 95)
features_df['ae_anomaly'] = mse > threshold
```

```python
import numpy as np

def sliding_window_anomaly(mags, window_size=5, threshold=3):
    anomalies = []
    for i in range(len(mags)):
        if i < window_size:
            anomalies.append(False)
        else:
            window = mags[i-window_size:i]
            mean = np.mean(window)
            std = np.std(window)
            if std == 0:
                anomalies.append(False)
            else:
                anomalies.append(abs(mags[i] - mean) > threshold * std)
    return anomalies

# Apply to each object in long_df
long_df = long_df.sort_values(['objectid', 'hmjd'])
long_df['sw_anomaly'] = False  # Initialize column

for oid in long_df['objectid'].unique():
    mask = long_df['objectid'] == oid
    mags = long_df.loc[mask, 'mag'].astype(float).values
    anomalies = sliding_window_anomaly(mags)
    long_df.loc[mask, 'sw_anomaly'] = anomalies
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_object_light_curve(objectid, long_df, results_df):
    """
    Plot a light curve for a specific object with anomaly detection markers
    from different algorithms.
    
    Parameters:
    objectid - The ID of the object to plot
    long_df - DataFrame with time series data (hmjd, mag, objectid)
    results_df - DataFrame with algorithm results per object
    """
    # Filter data for this object
    obj_data = long_df[long_df['objectid'] == objectid].sort_values('hmjd')
    
    # Get algorithm results for this object
    obj_results = results_df[results_df['objectid'] == objectid]
    
    # Create the plot
    plt.figure(figsize=(12, 6))
    
    # Plot the light curve
    plt.plot(obj_data['hmjd'], obj_data['mag'], 'k-o', label='Light Curve')
    
    # Plot markers for algorithm detections
    # For sliding window, mark points where anomalies were detected
    if 'sw_anomaly' in long_df.columns:
        anomaly_points = obj_data[obj_data['sw_anomaly']]
        if not anomaly_points.empty:
            plt.scatter(anomaly_points['hmjd'], anomaly_points['mag'], 
                      color='red', s=80, marker='o', label='Sliding Window')
    
    # For other algorithms, mark the earliest point if the object was flagged
    if 'rf_anomaly' in obj_results.columns and obj_results['rf_anomaly'].values[0]:
        earliest = obj_data.iloc[0]
        plt.scatter(earliest['hmjd'], earliest['mag'], 
                  color='blue', s=80, marker='x', label='Random Forest')
        
    if 'ae_anomaly' in obj_results.columns and obj_results['ae_anomaly'].values[0]:
        earliest = obj_data.iloc[0]
        plt.scatter(earliest['hmjd'], earliest['mag'], 
                  color='green', s=80, marker='s', label='Autoencoder')
    
    if 'if_anomaly' in obj_results.columns and obj_results['if_anomaly'].values[0]:
        earliest = obj_data.iloc[0]
        plt.scatter(earliest['hmjd'], earliest['mag'], 
                  color='purple', s=80, marker='^', label='Isolation Forest')
    
    if 'svm_anomaly' in obj_results.columns and obj_results['svm_anomaly'].values[0]:
        earliest = obj_data.iloc[0]
        plt.scatter(earliest['hmjd'], earliest['mag'], 
                  color='orange', s=80, marker='d', label='One-Class SVM')
    
    # Customize the plot
    plt.title(f'Object ID: {objectid}')
    plt.xlabel('Heliocentric MJD')
    plt.ylabel('Magnitude')
    plt.gca().invert_yaxis()  # Astronomical convention: brighter objects at top
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    
    return plt

# Example usage:
# For a single object
plot_object_light_curve(objectid=results_df['objectid'].iloc[0], 
                       long_df=long_df, 
                       results_df=results_df)
plt.show()

# To plot multiple objects (e.g., the top 5 objects with most algorithm flags)
top_objects = results_df.sort_values('n_algos_flagged', ascending=False).head(5)
for obj_id in top_objects['objectid']:
    plot_object_light_curve(obj_id, long_df, results_df)
    plt.savefig(f'lightcurve_{obj_id}.png', dpi=300)
    plt.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Ensure you have your long_df and results_df ready
# long_df: contains 'objectid', 'hmjd', 'mag', and 'sw_anomaly' (point-by-point flag)
# results_df: contains 'objectid', 'rf_anomaly', 'if_anomaly', 'svm_anomaly', 'ae_anomaly' (object-level flags)

def plot_object_light_curve_detailed(objectid, long_df, results_df):
    """
    Plot a light curve for a specific object with anomaly detection markers
    from different algorithms.
    
    - Sliding Window: Marks individual anomalous points.
    - Other Algorithms (RF, AE, IF, SVM): Marks all points if the object is flagged by them.
    
    Parameters:
    objectid (int): The ID of the object to plot.
    long_df (pd.DataFrame): DataFrame with time series data (hmjd, mag, objectid, sw_anomaly).
    results_df (pd.DataFrame): DataFrame with object-level anomaly flags (e.g., rf_anomaly, ae_anomaly).
    """
    
    obj_data = long_df[long_df['objectid'] == objectid].sort_values('hmjd')
    
    # Get object-level anomaly flags for this specific object
    # Use .iloc[0] to get the single row for this object
    obj_flags = results_df[results_df['objectid'] == objectid].iloc[0] 
    
    plt.figure(figsize=(12, 6))
    
    # Plot the main light curve (all points)
    plt.plot(obj_data['hmjd'], obj_data['mag'], 'k-', label='Light Curve', alpha=0.7)
    
    # Define markers and colors for each algorithm
    # These will be used if the *object* is flagged by the algorithm
    # and mark all points for that object for visual consistency
    algo_markers = {
        'rf_anomaly': {'color': 'blue', 'marker': 'x', 'label': 'Random Forest'},
        'ae_anomaly': {'color': 'green', 'marker': 's', 'label': 'Autoencoder'},
        'if_anomaly': {'color': 'purple', 'marker': '^', 'label': 'Isolation Forest'},
        'svm_anomaly': {'color': 'orange', 'marker': 'd', 'label': 'One-Class SVM'}
    }
    
    # Plot detections:
    # 1. Sliding Window (point-by-point anomaly detection)
    sw_anomaly_points = obj_data[obj_data['sw_anomaly']]
    if not sw_anomaly_points.empty:
        plt.scatter(sw_anomaly_points['hmjd'], sw_anomaly_points['mag'], 
                    color='red', s=80, marker='o', edgecolors='black', linewidth=0.8, label='Sliding Window', zorder=5)
    
    # 2. Other Algorithms (if the object is flagged, mark ALL its points with the algorithm's specific marker)
    for algo_col, props in algo_markers.items():
        # Check if the object was flagged by this algorithm
        if algo_col in obj_flags and obj_flags[algo_col]:
            # Plot all points of the light curve with this algorithm's marker
            plt.scatter(obj_data['hmjd'], obj_data['mag'], 
                        color=props['color'], s=50, marker=props['marker'], label=props['label'], zorder=4)

    # Customize the plot
    plt.title(f'Object ID: {objectid}', fontsize=16)
    plt.xlabel('Heliocentric MJD', fontsize=12)
    plt.ylabel('Magnitude', fontsize=12)
    plt.gca().invert_yaxis()  # Astronomical convention: brighter objects have lower magnitude
    plt.legend(loc='lower left', fontsize=10, bbox_to_anchor=(1.02, 0)) # Place legend outside
    plt.grid(True, alpha=0.3, linestyle='--')
    plt.tight_layout(rect=[0, 0, 0.85, 1]) # Adjust layout to prevent legend overlap
    
    return plt

# --- Example Usage ---

# Get a list of interesting object IDs to plot
# For instance, objects flagged by multiple algorithms (high confidence)
# Sort by the number of flags to pick the "most anomalous"
top_anomalous_objects_ids = results_df.sort_values(by='n_algos_flagged', ascending=False)['objectid'].head(5).tolist()

print(f"Plotting light curves for object IDs: {top_anomalous_objects_ids}")

# Loop through and plot each selected object
for obj_id in top_anomalous_objects_ids:
    plot_object_light_curve_detailed(obj_id, long_df, results_df)
    plt.savefig(f'light_curve_{obj_id}_detections.png', dpi=300) # Save the figure
    plt.show() # Display the figure
    plt.close() # Close the figure to free memory
```
